```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('grades.csv', sep = ';')
df.head()
```

```python
df[df["absence_status"] != "\\N"]
```

```python
pd.set_option('display.max_colwidth', 100)
pd.set_option('display.width', 200)

table = []
for column in df.columns[1:]:
    table.append(df[column].unique())

table_df = pd.DataFrame({
    'columns' : df.columns[1:],
    'unique values' : table
})

table_df
```

```python
table_df.to_csv('unique_values.csv', index=False, encoding='utf-8')
```
