```python
from pathlib import Path
import sys
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np

sys.path.append(str(Path().resolve().parent))
from src.data_processing.data_processing_copy import make_data_from_faculty, clean_data, make_data_from_program, make_features, process_data
from src.data_processing.data_processing_copy import main as main_data
```

```python
def str_to_int_stadent_status (df : pd.DataFrame):
    df["student_status"] = df["student_status"].map({
        "study": 0,
        "graduated": 0,
        "expelled": 1,
        "leave" :  1
    })
    return df
df_clean = str_to_int_stadent_status(df_clean)
```

```python


categorical_cols = ['place_type']

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_array = encoder.fit_transform(df_clean[categorical_cols])

encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(categorical_cols)
)

```

```python
df_final = pd.concat([df_clean.drop(columns=categorical_cols), encoded_df], axis=1)
df_final
```

```python
df_final.info()
```

```python
df_final['avr_4_grade'] = df_final.filter(like='3_4_Дисциплина:').mean(axis=1)
df_final.drop(columns=df_final.filter(like='3_4_Дисциплина:'))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X = df_final.drop(columns=['student_id_hash', 'program'])
y = df_final['avr_4_grade']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)


print (mean_absolute_error(y_test, y_pred))

# print(precision_score(y_test, y_pred))
```
